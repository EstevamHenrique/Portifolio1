**Após primeira analise foi percebido que apenas a variável S=(enxofre) tem informações faltantes, como não tenho dados para subistiur os faltantes opto por excluir essa variável e a varial ponto que se acemelha a um índice.**


```python
#excluindo variáveis usando .drop

df = df.drop('S', axis = 1)
df = df.drop('PONTO', axis = 1)
df.info()
```

vamos categorizar a variável argila(clay) entre, arenosa, média e argilosa criando uma função.

```python
def categoria_argila(row):
    if row['Clay'] < 150:
        return 'arenosa'
    elif row['Clay'] >= 150 and row['Clay'] <=350:
        return 'media'
    elif row['Clay'] >=350 and row['Clay'] <=650:
        return 'argilosa'
    return 'muito argilosa'

df['Clay_Categore'] = df.apply(lambda row: categoria_argila(row), axis = 1)
```

Aqui identificamos apenas duas categorias da argila(Clay)

```python
df['Clay_Categore'].value_counts()
```

Avaliando a produtividade em função da textura do solo

```python
from seaborn.categorical import boxplot


plt.subplots(figsize= (12,8))
sns.set(font_scale=1, style= 'white')
plt.rcParams.update({'font.family':'Arial'})

boxplot = sns.boxplot(x='Clay_Categore', y='Yield', data=df, palette='rocket_r', 
                      order= ['media','argilosa'], fliersize=3, linewidth=2, width=0.7)

boxplot.set_xlabel('Classe Textural do Solo', fontsize=18, labelpad= 10)
boxplot.set_ylabel('Produtividade (t/ha)', fontsize= 18, labelpad= 10)
boxplot.tick_params(labelsize=14)
plt.subplots_adjust(left=0.15, right=.9, bottom=0.1, top=0.9)
```

aqui podemos perceber que onde a produtividade é maior onde a textura do solo é mais argilosa, sendo assim podemos identificar quais talhões possuem taxas menores de argila e propor um tratamento.

**Início modelamento dos dados.**





```python
#importando biblioteca sklearn
from sklearn.model_selection import train_test_split

df_x = df.drop(['Yield'], axis= 1) # estou tirando a variável de produção e deixando apenas as variáveis de solo
df_y = df['Yield'] #apenas analise de produção

#dividindo as variáveis em treino e teste, vamos trabalhar com 70% treino e 30% teste
df_x_train, df_x_test, df_y_train, df_y_test = train_test_split(df_x, df_y, train_size=0.7, test_size=0.3)
```

```python
df_x_test
```

    **Normalizando/ Padronizando as variáveis numéricas**
Quando falamos de variáveis numéricas temos que padronizar/normalizar, pegar o valor maximo da variavel chamar de 1 e o valor minimo chamar de 0, assim igualamos os pesos das variáveis.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_x_train[['X (m)', 'Y (m)', 'Elevation (m)', 'Slope (%)','Clay', 'Sand', 'Silt', 'OM', 'pH', 'P', 'K', 'Ca', 'Mg', 'H+Al', 'SEB',
       'CEC', 'BS', 'B', 'Cu', 'Fe', 'Mn']] = scaler.fit_transform(df_x_train[['X (m)', 'Y (m)', 'Elevation (m)', 'Slope (%)','Clay', 'Sand', 'Silt', 'OM', 'pH', 'P', 'K', 'Ca', 'Mg', 'H+Al', 'SEB',
       'CEC', 'BS', 'B', 'Cu', 'Fe', 'Mn']])
df_x_test[['X (m)', 'Y (m)', 'Elevation (m)', 'Slope (%)','Clay', 'Sand', 'Silt', 'OM', 'pH', 'P', 'K', 'Ca', 'Mg', 'H+Al', 'SEB',
       'CEC', 'BS', 'B', 'Cu', 'Fe', 'Mn']] = scaler.fit_transform(df_x_test[['X (m)', 'Y (m)', 'Elevation (m)', 'Slope (%)','Clay', 'Sand', 'Silt', 'OM', 'pH', 'P', 'K', 'Ca', 'Mg', 'H+Al', 'SEB',
       'CEC', 'BS', 'B', 'Cu', 'Fe', 'Mn']])
```

Agora todas as variáveis vão de 0 a 1

```python
df_x_train
```

**Para as variaveis categorias precisamos transformalas em numéricas, criaremos dumies.**

```python
df_x_train = pd.get_dummies(df_x_train, drop_first=True)
df_x_test = pd.get_dummies(df_x_test, drop_first=True)
```

```python
df_x_train
```

# Aplicando Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=100)

rfr.fit(df_x_train, df_y_train)
```

#Predizendo na base de test


```python
rfr_pred = rfr.predict(df_x_test)
```

# Avaliando as métricas (R2) accuracy

```python
from sklearn.metrics import r2_score
 
r2_score(df_y_test, rfr_pred)
```

# Avaliando as métricas (RMSE) 

```python
from sklearn.metrics import mean_squared_error 

mean_squared_error(df_y_test, rfr_pred, squared=False)
```

# Avaliando o impacto das variáveis top 10

```python
importances = rfr.feature_importances_
std = np.std([tree.feature_importances_ for tree in rfr.estimators_], axis=0)

rfc_importances = pd.Series(importances, index=df_x_train.columns)
imp = rfc_importances.sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(15,8))

imp.plot.bar(yerr=std, ax=ax)

ax.set_title("Importancia dos recursos usando MDI")
ax.set_ylabel("Redução média na impureza")

plt.xticks(fontsize=15)

plt.xlim([-1, 9.5])

fig.tight_layout()
```

```python
lw = 1.5

fig, ax = plt.subplots(1, 1,figsize=(12, 8))

sns.scatterplot(x=rfr_pred, y=df_y_test, data=df_x_test)

plt.xlabel('y_Predict')
plt.ylabel('y_True')
plt.title('Random Forest Regression')

plt.legend()

plt.show()
```

# Ajustando hiperparâmetros

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'bootstrap': [True, False], 
              'max_depth': [10, 50, 100, None], 
              'max_features': ['auto', 'sqrt'],
              'n_estimators': [100, 500, 1000, 1500]}
```

```python
grid = GridSearchCV(estimator=rfr, param_grid=param_grid, cv=3)
```

```python
grid.fit(df_x_train, df_y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(df_x_test)
```

```python
r2_score(df_y_test, grid_predictions)
```

```python
mean_squared_error(df_y_test, grid_predictions, squared=False)
```

Após simular os hiper parametros obtemos uma pequena melhora apenas na métrica de (RMSE), diminuindo a perca.
